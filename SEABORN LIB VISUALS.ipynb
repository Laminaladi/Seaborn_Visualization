```python
#seaborn is also visualiztion library and it uses the matplotlib background 
#importing the libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
datas=pd.read_excel("products summary.xlsx",sheet_name="sales details")
datas
```

```python
plt.figure(figsize=(11,5))
sns.barplot(x="SALES RAP",y="UNIT PRICE",data=datas)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="SALES RAP",y="UNIT PRICE",data=datas).set_title("UNIT PRICE BY SALES RAP")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.stripplot(x="PRODUCT",y="TOTAL",data=datas,hue="PRODUCT").set_title("TOTAL SALES BY PRODUCT")
plt.xlabel("PRODUCT")
plt.ylabel("TOTAL")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x="TOTAL",y="ZONE",data=datas).set_title("TOTAL SALES BY ZONE")
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(x="MONTH",y="TOTAL",data=datas).set_title("TOTAL SALES BY MONTH")
plt.show()
```

```python
""""with subplot () function you can draw multiple plots in one figure
The subplot() function takes three arguments that describes the layout of the figure.
The layout is organized in rows and columns, which are represented by the first
and second argument.
The third argument represents the index of the current plot. plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
pit.subplot(1, 2, 2)
the figure has 1 row, 2 columns, and this plot is the second plot."""

plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
sns.barplot(x="SALES RAP",y="UNIT PRICE",data=datas)

plt.subplot(2,2,2)
sns.barplot(x="SALES RAP",y="UNIT PRICE",data=datas).set_title("UNIT PRICE BY SALES RAP")

plt.subplot(2,2,3)
sns.barplot(x="TOTAL",y="ZONE",data=datas).set_title("TOTAL SALES BY ZONE")

plt.subplot(2,2,4)
sns.stripplot(x="PRODUCT",y="TOTAL",data=datas,hue="PRODUCT").set_title("TOTAL SALES BY PRODUCT")


plt.show()
```
